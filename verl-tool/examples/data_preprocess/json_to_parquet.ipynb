```python
import pandas as pd
import json
import random
import os
def json_to_parquet_safe(json_file: str, parquet_file: str):
    # 先用 json.load 读入
    file_name = os.path.basename(json_file).split('.')[0]
    with open(json_file, "r", encoding="utf-8") as f:
        try:
            data = json.load(f)  # 适合数组形式的 JSON
        except json.JSONDecodeError:
            # 如果是 JSON lines 格式
            data = [json.loads(line) for line in f]
    # data = random.sample(data, 100)
    # train_data = data[:-100]
    # val_data = data[-100:]
    # 转 DataFrame（自动填充缺失字段为 NaN）
    # df_train = pd.DataFrame(train_data)
    # df_val = pd.DataFrame(val_data)
    df = pd.DataFrame(data)
    # train_path = f'{parquet_file}/sample_300_train.parquet'
    # val_path = f'{parquet_file}/sample_300_val.parquet'
    # 保存 parquet
    # df_train.to_parquet(train_path, engine="pyarrow", index=False)
    # df_val.to_parquet(val_path, engine="pyarrow", index=False)
    df.to_parquet(f'{parquet_file}/{file_name}.parquet', engine="pyarrow", index=False)

    print(f"转换完成: {json_file} -> {parquet_file}")



if __name__ == "__main__":
    json_to_parquet_safe("/home/QiuJihao/Workspace/RL/latest_code/qwen3_code/train_data/ori_data_not_processed/cgbench_sample_800_video_rl_use.json", "/home/QiuJihao/Workspace/RL/latest_code/qwen3_code/train_data/ori_data_not_processed")
```
